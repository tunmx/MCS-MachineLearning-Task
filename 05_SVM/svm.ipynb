# Import Package

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from mpl_toolkits.mplot3d import Axes3D
```

## Load Wine Dataset

```python
wine = datasets.load_wine()
X = wine.data
y = wine.target
feature_names = wine.feature_names
target_names = wine.target_names
```

## Data To DataFrame

```python
df = pd.DataFrame(X, columns=feature_names)
df['target'] = y

print("Overview of the data set：")
display(df.head())

print("\nData statistics：")
display(df.describe())

print("\ncategorical distribution：")
sns.countplot(x='target', data=df)
plt.title('Wine Classes Distribution')
plt.show()
```

## Normalization

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Split Dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

## Build SVM

```python
# Initializing and training SVM classifiers (linear kernel)
svm_classifier = SVC(kernel='linear', C=1.0, random_state=42)
svm_classifier.fit(X_train, y_train)
```

## Prediction and Evaluation

```python
# prediction
y_pred = svm_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy of linear kernel SVM model: {accuracy}")
print("\nClassified report:")
print(classification_report(y_test, y_pred))
```

## Visual Report

```python
# confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=target_names, yticklabels=target_names)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
# 3D visualization of classification results
def plot_3d(X, y, model, title):
    pca = PCA(n_components=3)
    X_reduced = pca.fit_transform(X)

    fig = plt.figure(figsize=(10, 7))  # 设置图形大小
    ax = fig.add_subplot(111, projection='3d')

    # Plot a scatter plot of the training set
    scatter = ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y, cmap='viridis', edgecolor='k', s=40)

    # Draw a classification hyperplane
    xx, yy = np.meshgrid(np.linspace(X_reduced[:, 0].min(), X_reduced[:, 0].max(), num=50),
                         np.linspace(X_reduced[:, 1].min(), X_reduced[:, 1].max(), num=50))
    zz = (-model.intercept_[0] - model.coef_[0][0] * xx - model.coef_[0][1] * yy) / model.coef_[0][2]
    ax.plot_surface(xx, yy, zz, alpha=0.2, color='blue')

    ax.set_title(title)
    ax.set_xlabel('PCA 1')
    ax.set_ylabel('PCA 2')
    ax.set_zlabel('PCA 3')
    legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
    ax.add_artist(legend1)
    plt.show()


plot_3d(X_train, y_train, svm_classifier, "3D Visualization of SVM Classification on Wine Data (Training Set)")
```

## Try Other Kernel Functions and Compare Them

We will evaluate the SVM classifier with different kernels ('linear', 'poly', 'rbf', 'sigmoid') to compare their performance.

```python
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
results = []

for kernel in kernels:
    svm_classifier = SVC(kernel=kernel, C=1.0, random_state=42)
    svm_classifier.fit(X_train, y_train)
    y_pred = svm_classifier.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    results.append((kernel, accuracy))
    print(f"\nKernel: {kernel}")
    print(f"Accuracy: {accuracy}")
    print(classification_report(y_test, y_pred))
    
    # Visual confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', xticklabels=target_names, yticklabels=target_names)
    plt.title(f'Confusion Matrix - {kernel} Kernel')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
```

## Result

```python
# summarized results
results_df = pd.DataFrame(results, columns=['Kernel', 'Accuracy'])
print("\nThe results of different kernel functions are compared：")
display(results_df)
```

```python
sns.barplot(x='Kernel', y='Accuracy', data=results_df)
plt.title('SVM Kernel Comparison')
plt.ylim(0.8, 1.0)
plt.show()
```

## Conclusion
The comparison of different kernel functions for the SVM classifier on the Wine dataset indicates that the linear and RBF (Radial Basis Function) kernels achieve the highest accuracy, both scoring approximately 0.981. The polynomial and sigmoid kernels show slightly lower performance, with accuracies of approximately 0.963. This suggests that for the Wine dataset, linear and RBF kernels are more effective in distinguishing between the different classes, making them the preferred choices for this classification task.
