### 3. KNN classification 

We designed a KNN classifier to categorize the MNIST dataset, and conducted experiments using both L1 and L2 distance metrics.

```python
import numpy as np
from tqdm import tqdm

class KNN:
    def __init__(self, k=3, loss='L2'):
        self.k = k
        if loss == 'L1':
            self.loss_func = self.manhattan_distance
        elif loss == 'L2':
            self.loss_func = self.euclidean_distance
        else:
            raise NotImplemented

    def fit(self, X, y):
        self.X_train = X
        self.y_train = np.asarray(y)
    
    @staticmethod
    def euclidean_distance(a, b):
        return np.sqrt(np.sum((a - b) ** 2, axis=1))
    
    @staticmethod
    def manhattan_distance(a, b):
        return np.sum(np.abs(a - b), axis=1)
    
    @staticmethod
    def calculation_acc(y, y_hat):
        accuracy = np.mean(y_hat == y_test)
        return accuracy

    def predict(self, X_test):
        y_pred = []
        for x in tqdm(X_test):
            # Calculate the distance between the test sample and all the training samples
            distances = self.loss_func(x, self.X_train)
            # The index of the nearest k samples is obtained
            nearest_indices = np.argsort(distances)[:self.k]
            # Get the labels of these k samples
            nearest_labels = self.y_train[nearest_indices]
            # The predictive labels of the test sample are determined according to the voting mechanism
            pred_label = np.argmax(np.bincount(nearest_labels))
            y_pred.append(pred_label)
        return np.array(y_pred)
    
    def evaluate(self, X_test, y_test):
        y_pred = self.predict(X_test)
        accuracy = np.mean(y_pred == y_test)
        return accuracy
```

#### Select K value

We select a set of k values to validate the MNIST dataset, following the previously completed data preprocessing steps. Using the KNN classifier, we classify and evaluate the accuracy, aiming to identify the most suitable k value.

```python
from sklearn.model_selection import KFold
import numpy as np
import matplotlib.pyplot as plt

def knn_k_steps(X_train, y_train, X_test, y_test, k_values, max_test_samples=500, loss='L2'):
    """
    The knn classification is verified using a set of k values, tests each k value for accuracy on the test set.
    """
    best_accuracy = 0
    best_k = None
    accuracies = []

    # Ensure that the test data does not exceed the number specified by max test samples
    X_test_limited = X_test[:min(len(X_test), max_test_samples)]
    y_test_limited = y_test[:min(len(y_test), max_test_samples)]
    
    for k in k_values:
        knn = KNN(k=k, loss=loss)
        knn.fit(X_train, y_train)
        
        accuracy = knn.evaluate(X_test_limited, y_test_limited)
        accuracies.append(accuracy)

        print(f"Accuracy with k={k}: {accuracy}")
        
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_k = k

    print(f"Best k value: {best_k} with accuracy: {best_accuracy}")

    return best_k, accuracies

def plot_k_vs_accuracy(k_values, accuracies):
    accuracy_values = [accuracies[idx] for idx, k in enumerate(k_values)]
    
    plt.figure(figsize=(10, 6))
    plt.plot(k_values, accuracy_values, marker='o', linestyle='-', color='b')
    plt.title('KNN Accuracy vs. K Value')
    plt.xlabel('K Value')
    plt.ylabel('Accuracy')
    plt.xticks(k_values)
    plt.grid(True)
    plt.show()
```

#### Use Euclidean distance
Evaluate the MNIST dataset with a set of k values using the Euclidean metric.

```python
# k steps: 3, 5, 7, 9
k_values = list(range(3, 11, 2))
# Save time extracting only a small portion of the test set data
max_test_samples = 250
# Use Euclidean distance
loss = 'L2'
best_k, accuracies = knn_k_steps(X_train, 
                                 y_train, 
                                 X_test, 
                                 y_test, 
                                 k_values, 
                                 max_test_samples=max_test_samples, 
                                 loss=loss)
# displaying results
plot_k_vs_accuracy(k_values, accuracies)
```

#### Use Manhattan distance

Evaluate the MNIST dataset with a set of k values using the Manhattan metric.

```python
# k steps: 3, 5, 7, 9
k_values = list(range(3, 11, 2))
# Save time extracting only a small portion of the test set data
max_test_samples = 250
# Use Manhattan distance
loss = 'L1'
best_k, accuracies = knn_k_steps(X_train, 
                                 y_train, 
                                 X_test, 
                                 y_test, 
                                 k_values, 
                                 max_test_samples=max_test_samples, 
                                 loss=loss)
# displaying results
plot_k_vs_accuracy(k_values, accuracies)
```

### Analyze the results

From the analysis of the experimental results above, we can observe that when experimenting with a specified number of test subsets, the best accuracy is achieved with a k value of **3** and by using the **Euclidean distance** for measurement. We use the optimal parameters found in the current test to build a KNN model, and then perform inference on a batch of sub-test sets to obtain results for analysis.

```python
import random

def select_random_subset(X, y, num=500):
    """A subset was randomly drawn from the MNIST dataset."""
    indices = random.sample(range(len(X)), num)
    
    X_subset = X[indices]
    y_subset = y[indices]
    
    return X_subset, y_subset
```

Build a knn model

```python
# Build a knn model
knn = KNN(k=3, loss="L2")
knn.fit(X_train, y_train)

# Select a sub test datasets
X_sub_test, y_sub_test = select_random_subset(X_test, y_test, 1000)
print(f"X_sub_test: {X_sub_test.shape}")
print(f"y_sub_test: {y_sub_test.shape}")
```

Perform knn model inference.

```python
y_hat_sub_test = knn.predict(X_sub_test)
accuracy = np.mean(y_sub_test == y_hat_sub_test)
print(f"X_sub_test accuracy: {accuracy}")
```

We use a confusion matrix to analyze predictions on a test subset, aiming to visually identify misclassification patterns and improve model accuracy.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

def plot_confusion_matrix(y, y_hat):
    """
    Plot the confusion matrix for the MNIST dataset.
    """
    cm = confusion_matrix(y, y_hat)
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=np.arange(10), yticklabels=np.arange(10))
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix for MNIST')
    plt.show()

plot_confusion_matrix(y_sub_test, y_hat_sub_test)
```

Analyzing the test results, it's easy to notice that the error rates between 8 and 3, as well as 9 and 3, are relatively high, indicating a propensity for misclassification during the prediction process. We can employ visualization techniques to display the images that were incorrectly predicted, offering a visual perspective to observe the data that led to confusion.

```python
import matplotlib.pyplot as plt
import numpy as np

def show_misclassified_images(X, y, y_hat, num_images=10):
    """
    An image showing the prediction error and its true and predicted values.
    """
    misclassified_indices = [i for i, (true, pred) in enumerate(zip(y, y_hat)) if true != pred]
    
    num_images = min(len(misclassified_indices), num_images) 
    num_rows = np.ceil(num_images / 6).astype(int)
    
    plt.figure(figsize=(15, 2.5 * num_rows))  
    for i, index in enumerate(misclassified_indices[:num_images]):
        plt.subplot(num_rows, 6, i + 1)  
        image = X[index].reshape(28, 28) 
        plt.imshow(image, cmap='gray')
        plt.title(f"True: {y[index]}\nPred: {y_hat[index]}", color='red')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

show_misclassified_images(X_sub_test, y_sub_test, y_hat_sub_test, num_images=24)
```

#### Analysis of Causes

- Visual Similarity: The confusion between the digits 8, 3, and 9 in KNN classification can arise from their visual resemblance. Both 8 and 3 involve curves and loops, and 9 can be seen as a variant of 3, leading to close proximity in feature space and potential misclassification.

- Feature Representation Limitations: The performance of KNN heavily relies on the representation of data. Using raw pixel values as features might not capture all necessary details to distinguish between visually similar digits effectively.

- Noise and Deformation: Variability in handwriting, such as noise, skew, and deformation, complicates classification. An inappropriate k value might exacerbate these issues by being too sensitive to noise (small k) or failing to adequately define class boundaries (large k).
